```python
#指数関数
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-4,4,100)#-4から４までを１００等分した等差数列を生成
y1=2**x
y2=3**x
y3=0.5**x

plt.plot(x,y1,'black',label='$y=2^x$')
plt.plot(x,y2,'blue',label='$y=3^x$')
plt.plot(x,y3,'gray',label='$y=0.5^x$')

plt.ylim(-2,6)
plt.xlim(-4,4)
plt.grid(True)
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1=np.linspace(-8,8,100)
y1=2**x1

x2=np.linspace(0.001,8,100)#np.log(0)はエラーになるのでｘ２に０を含めない
y2=np.log(x2)/np.log(2)#底が２、指数がｘ２のときの対数関数を計算
#log(2)x=log(x)/log(2)

plt.plot(x1,y1,'black',label='$y=2^x$')
plt.plot(x2,y2,'blue',label='$y=log(2)x$')
plt.plot(x1,x1,'black',label='$y=x$',linestyle='dashed')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.grid(True)
plt.legend(loc='lower right')
plt.show()
```

```python
#指数関数の微分
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-5,5,100)
a=int(2)
y=a**x
dy=np.log(a)*y

plt.plot(x,y,'black',label='$y=%d^x$'%a)
plt.plot(x,dy,'black',label='$dy/dx=%d^xlog(%d)$'%(a,a),linestyle='dashed')
plt.xlim(-5,5)
plt.ylim(-1,8)
plt.grid(True)
plt.legend(loc='lower right')
plt.show()
```

```python
#対数関数の微分
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0.0001,5,100)
y=np.log(x)
dy=1/x

plt.plot(x,y,'black',label='y=log(x)')
plt.plot(x,dy,'black',linestyle='--',label='dy/dx=1/x')
plt.xlim(-1,5)
plt.ylim(-8,8)
plt.grid(True)
plt.legend(loc='lower right')
plt.show()
```

```python
#シグモイド関数の微分
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-10,10,100)
y=1/(1+np.exp(-x))
dy=y*(1-y)

plt.plot(x,y,'black',label='$y=1/(1+e^{-x})$')
plt.plot(x,dy,'black',label='dy/dx=y(1-y)',linestyle='--')
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
```
