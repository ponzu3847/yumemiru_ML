```python
#数値微分を実行してグラフを表示する関数
import numpy as np
import matplotlib.pyplot as plt

def differential(f,x):
    '''
    数値微分を行う高階関数
    
    パラメータ
    -------------
    f : 関数オブジェクト
        数値微分に用いる関数
    x : int
        f(x)のxの値
    戻り値
    '''
    h=1e-4
    return (f(x+h)-f(x-h))/(2*h)

def function(x):
    '''
    数値微分で使用する関数
    
    パラメータ
    ---------------
    x : int
        f(x)のｘの値
    戻り値
    '''
    return 0.01*x**2+0.1*x

def draw_line(f,x):
    '''
    数値微分の値を傾きとする直線をプロットするラムダ式を生成する関数
    differentialを実行する
    
    パラメータ
    -----------------
    f : 関数オブジェクト
        数値微分に使用する関数
    x : int
        f(x)のｘの値
    戻り値
    -----------------
    lambdaオブジェクト
        数値微分の値を傾きとする直線をプロットするためのラムダ式
    '''
    dff=differential(f,x)#接線の傾きを求める
    print(dff)
    y=f(x)-dff*x#接線の傾きからy切片を求める
    return lambda n: dff*n+y#接線の関数を返す
    
###実行ブロック###
if __name__=='__main__':
    
    x=np.arange(0.0,20.0,0.1)
    y=function(x)
    
    plt.xlabel('x')
    plt.ylabel('f(x)')
    
    tf=draw_line(function,5)
    y2=tf(x)
    
    plt.plot(x,y)
    plt.plot(x,y2)
    plt.show()
    
    
```

```python
#勾配法を使って最適化を行う
import numpy as np
import matplotlib.pyplot as plt

def standardize(x):
    '''
    データの標準化を行う
    
    パラメータ
        x : 標準化前のx
    '''
    x_mean=x.mean()
    x_std=x.std()
    return (x-x_mean)/x_std
    
def f(standardized_x,parameter_0,parameter_1):
    '''
    回帰直線を求める関数
    
    パラメータ
        standardized_x : 標準化したx
        parameter_0 : パラメータφ0
        parameter_1 : パラメータφ1
    '''
    return parameter_0+parameter_1*standardized_x

def objective_f(standardized_x,y,parameter_0,parameter_1):
    '''
    目的関数(誤差関数)
    
    パラメータ
        standardized_x : 標準化したx
        y : yの値
        parameter_0 : パラメータφ0
        parameter_1 : パラメータφ1
    '''
    return 0.5*np.sum((y-f(standardized_x,parameter_0,parameter_1))**2)

def gradient_method(standardized_x,y):
    '''
    勾配法で最適化を行う
    
    パラメータ
        standardized_x : 標準化したx
        ｙ : yの値
    '''
    #パラメータを初期化
    parameter_0=np.random.rand()
    parameter_1=np.random.rand()
    
    #学習率をセット
    LNR=1e-3
    
    #パラメータ更新前後の目的関数の値の差を保持する変数を1で初期化
    difference=1
    
    #更新回数をカウントする変数を0で初期化
    count=0
    
    #パラメータの初期値を使用した目的関数の値をbeforeに代入
    before=objective_f(standardized_x,y,parameter_0,parameter_1)
    
    #目的関数との差分を保持するリスト
    differences=[]
    
    #誤差の差分が一定以下になるまでパラメータ更新を繰り返す
    while difference >1e-2:
        #φ0を更新する
        tmp_parameter_0=parameter_0-LNR*np.sum(f(standardized_x,
                                                 parameter_0,
                                                 parameter_1)-y)
        
        #φ1を更新する
        tmp_parameter_1=parameter_1-LNR*np.sum((f(standardized_x,
                                                  parameter_0,
                                                  parameter_1)-y)
                                               *standardized_x)
        
        #更新後のφ0、φ1の値をパラメータ用の変数に代入
        parameter_0=tmp_parameter_0
        parameter_1=tmp_parameter_1
        
        #更新後のパラメータで目的関数の値を求める
        after=objective_f(standardized_x,y,parameter_0,parameter_1)
        
        #更新前後の目的関数の値の差を求める
        difference=before-after
        
        #更新後の目的関数の値を更新前の変数に代入
        before=after
        
        #カウンター変数を1増やす
        count+=1
        
        #処理結果を表示
        log='({}) φ0: {:.3f} φ1: {:.3f} error: {:.4f}'
        print(log.format(count,parameter_0,parameter_1,difference))
        
        differences.append(difference)
    
    #グラフ表示用にパラメータ更新回数分の等差数列を作成
    x_count=np.arange(len(differences))
    plt.plot(x_count,differences)
    plt.grid('True')
    plt.show()
    
    #勾配法で求めたパラメータφ0、φ１の値を戻り値として返す
    return parameter_0,parameter_1

###実行ブロック###
if __name__=='__main__':
    
    data=np.loadtxt(fname='access.csv',
                   dtype='int',
                   delimiter=',',
                   skiprows=1)
    
    x=data[:,0]
    y=data[:,1]
    
    #xを標準化
    standardized_x=standardize(x)
    
    #勾配法でパラメータφ0、φ１の値を求める
    parameter_0,parameter_1=gradient_method(standardized_x,y)
    
    #グラフ表示用のx軸の値として等差数列を作成
    x_axis=np.linspace(-3,3,100)
    plt.plot(standardized_x,y,'o')
    
    #回帰直線をプロット
    plt.plot(x_axis,f(x_axis,parameter_0,parameter_1))
    
    plt.grid('True')
    plt.show()

input_x=input('予測に使用するxの値を入力してください>')
x_mean=x.mean()
std=x.std()
pred=(int(input_x)-x_mean)/std
print(f(pred,parameter_0,parameter_1))
```

```python
#勾配法を使って多項式回帰の最適化を行う
import numpy as np
import matplotlib.pyplot as plt

def standardize(x):
    '''
    データの標準化を行う
    
    パラメータ
        x : 標準化前のx
    '''
    x_mean=x.mean()
    x_std=x.std()
    return (x-x_mean)/x_std
    
def create_matrix(standardized_x):
    '''
    すべての要素が1のベクトル、standardized_x、standardized_xの2乗を列方向に連結して行列を作成し、これを転置して返す
    
    パラメータ
        standardized_x : 標準化したx
    '''
    return np.vstack(
        [np.ones(standardized_x.size),
        standardized_x,
        standardized_x**2
        ]).T

def f(matrix_x,parameter):
    '''
    多項式回帰
    
    パラメータ
        matrix_x : xの行列
        parameter : パラメータφ、φ1、φ2の行ベクトル
    '''
    return np.dot(matrix_x,parameter)

def objective_f(matrix_x,y,parameter):
    '''
    目的関数(誤差関数)
    
    パラメータ
        matrix_x : xの行列
        y : yの値
        parameter : パラメータφ、φ1、φ2の行ベクトル
    '''
    return 0.5*np.sum((y-f(matrix_x,parameter))**2)

def polynominal_regression(matrix_x,y):
    '''
    多項式回帰で最適化を行う
    
    パラメータ
        matrix_x : xの行列
        ｙ : yの値
    '''
    #パラメータを初期化
    parameter=np.random.rand(3)
    
    #学習率をセット
    LNR=1e-3
    
    #パラメータ更新前後の目的関数の値の差を保持する変数を1で初期化
    difference=1
    
    #更新回数をカウントする変数を0で初期化
    count=0
    
    #パラメータの初期値を使用した目的関数の値をbeforeに代入
    before=objective_f(matrix_x,y,parameter)
    
    #目的関数との差分を保持するリスト
    differences=[]
    
    #誤差の差分が一定以下になるまでパラメータ更新を繰り返す
    while difference >1e-2:
        #パラメータを更新する
        parameter=parameter-LNR*np.dot(f(matrix_x,parameter)-y,matrix_x)
        
        #更新後のパラメータで目的関数の値を求める
        after=objective_f(matrix_x,y,parameter)
        
        #更新前後の目的関数の値の差を求める
        difference=before-after
        
        #更新後の目的関数の値を更新前の変数に代入
        before=after
        
        #カウンター変数を1増やす
        count+=1
        
        #処理結果を表示
        log='({}) parameter: {} error: {:.4f}'
        print(log.format(count,parameter,difference))
        
        differences.append(difference)
    
    #グラフ表示用にパラメータ更新回数分の等差数列を作成
    x_count=np.arange(len(differences))
    plt.plot(x_count,differences)
    plt.grid('True')
    plt.show()
    
    #勾配法で求めたパラメータφ0、φ１の値を戻り値として返す
    return parameter

###実行ブロック###
if __name__=='__main__':
    
    data=np.loadtxt(fname='access.csv',
                   dtype='int',
                   delimiter=',',
                   skiprows=1)
    
    x=data[:,0]
    y=data[:,1]
    
    #xを標準化
    standardized_x=standardize(x)
    
    #標準化したｘを行列にする
    matrix_x=create_matrix(standardized_x)
    
    #多項式回帰でパラメータの値を求める
    parameter=polynominal_regression(matrix_x,y)
    
    #グラフ表示用のx軸の値として等差数列を作成
    x_axis=np.linspace(-3,3,100)
    
    plt.plot(standardized_x,y,'o')
    plt.ylim(300,2500)
    
    #回帰曲線をプロット
    plt.plot(x_axis,f(create_matrix(x_axis),parameter))
    
    plt.grid('True')
    plt.show()
    
input_x=input('予測に使用するxの値を入力してください>')
x_mean=x.mean()
std=x.std()
pred=(int(input_x)-x_mean)/std
print(parameter[0]+parameter[1]*pred+parameter[2]*pred**2)
```

```python
#勾配法を使って多項式回帰の最適化を行う
import numpy as np
import matplotlib.pyplot as plt

def standardize(x):
    '''
    データの標準化を行う
    
    パラメータ
        x : 標準化前のx
    '''
    x_mean=x.mean()
    x_std=x.std()
    return (x-x_mean)/x_std

def create_matrix(standardized_x):
    '''
    すべての要素が1のベクトル、standardized_x、standardized_xの2乗を列方向に連結して行列を作成し、これを転置して返す
    
    パラメータ
        standardized_x : 標準化したx
    '''
    return np.vstack(
        [np.ones(standardized_x.size),
        standardized_x,
        standardized_x**2
        ]).T

def f(matrix_x,parameter):
    '''
    多項式回帰
    
    パラメータ
        matrix_x : xの行列
        parameter : パラメータφ、φ1、φ2の行ベクトル
    '''
    return np.dot(matrix_x,parameter)

def MSE(matrix_x,y,parameter):
    '''
    平均二乗誤差を求める
    
    パラメータ
        matrix_x : xの行列
        y : yの値
        parameter : パラメータφ、φ1、φ2の行ベクトル
    '''
    return (1/matrix_x.shape[0])*np.sum((y-f(matrix_x,parameter))**2)

def SGD(matrix_x,y):
    '''
    確率的勾配降下法で最適化を行う
    
    パラメータ
        matrix_x : xの行列
        ｙ : yの値
    '''
    #パラメータを初期化
    parameter=np.random.rand(3)
    
    #学習率をセット
    LNR=1e-3
    
    #パラメータ更新前後の目的関数の値の差を保持する変数を1で初期化
    difference=1
    
    #更新回数をカウントする変数を0で初期化
    count=0
    
    #パラメータの初期値を使用して平均二乗誤差を求める
    before=MSE(matrix_x,y,parameter)
    
    #目的関数との差分を保持するリスト
    differences=[]
    
    #誤差の差分が一定以下になるまでパラメータ更新を繰り返す
    while difference >1e-2:
        #学習データを並べ替えるためのランダムな順列を生成
        #配列pの要素をｘ、ｙのペアを作成する際のインデックスとして使用する
        p=np.random.permutation(matrix_x.shape[0])
        #学習データからｐのインデックスに従ってｘとｙのペアを1セットずつ取り出す
        for tmp_x,tmp_y in zip(matrix_x[p,:],y[p]):
            #学習データ一つを使ってパラメータを更新する
            parameter=parameter-LNR*np.dot(f(tmp_x,parameter)-tmp_y,tmp_x)
            
        #更新後のパラメータで平均二乗誤差を求める
        after=MSE(matrix_x,y,parameter)
        
        #更新前後の目的関数の値の差を求める
        difference=before-after
        
        #更新後の目的関数の値を更新前の変数に代入
        before=after
        
        #カウンター変数を1増やす
        count+=1
        
        #処理結果を表示
        log='({}) parameter: {} error: {:.4f}'
        print(log.format(count,parameter,difference))
        
        differences.append(difference)
    
    #グラフ表示用にパラメータ更新回数分の等差数列を作成
    x_count=np.arange(len(differences))
    plt.plot(x_count,differences)
    plt.grid('True')
    plt.show()
    
    #勾配法で求めたパラメータφ0、φ１の値を戻り値として返す
    return parameter

###実行ブロック###
if __name__=='__main__':
    
    data=np.loadtxt(fname='access.csv',
                   dtype='int',
                   delimiter=',',
                   skiprows=1)
    
    x=data[:,0]
    y=data[:,1]
    
    #xを標準化
    standardized_x=standardize(x)
    
    #標準化したｘを行列にする
    matrix_x=create_matrix(standardized_x)
    
    #確率的勾配降下法でパラメータの値を求める
    parameter=SGD(matrix_x,y)
    
    #グラフ表示用のx軸の値として等差数列を作成
    x_axis=np.linspace(-3,3,100)
    
    plt.plot(standardized_x,y,'o')
    plt.ylim(300,2500)
    
    #回帰曲線をプロット
    plt.plot(x_axis,f(create_matrix(x_axis),parameter))
    
    plt.grid('True')
    plt.show()
    
input_x=input('予測に使用するxの値を入力してください>')
x_mean=x.mean()
std=x.std()
pred=(int(input_x)-x_mean)/std
print(parameter[0]+parameter[1]*pred+parameter[2]*pred**2)
```
