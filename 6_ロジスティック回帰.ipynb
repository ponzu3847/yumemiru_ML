```python
#ロジスティック回帰で最適化を行う
import numpy as np
import matplotlib.pyplot as plt

def standardize(x):
    '''
    データを標準化する
    '''
    x_mean=x.mean(axis=0)
    std=x.std(axis=0)
    return (x-x_mean)/std

def sigmoid(x,parameter):
    return 1/(1+np.exp(-np.dot(x,parameter)))

def create_matrix(x):
    x0=np.ones([x.shape[0],1])#x.shape[0] 配列ｘの行数
    return np.hstack([x0,x])#横方向に行列を連結する [1,x1,x2]

def logistic_regression(matrix_x,t):
    '''
    ロジスティック回帰で最適化を行う
    '''
    parameter=np.random.rand(3)
    
    LNR=1e-3
    loop=3000
    count=0
    
    for i in range(loop):
        parameter=parameter-LNR*(np.dot((sigmoid(matrix_x,parameter)-t),matrix_x))
        count+=1
        print('({}) parameter: {}'.format(count,parameter))
    
    return parameter
    
###実行ブロック###
if __name__=='__main__':
    data=np.loadtxt('classification.csv',dtype='int',delimiter=',',skiprows=1)
    x=data[:,0:2]
    t=data[:,2]
    
    #標準化
    standardized_x=standardize(x)
    
    #行列ｘの生成
    matrix_x=create_matrix(standardized_x)
    
    #学習
    parameter=logistic_regression(matrix_x,t)
    
    #グラフの表示
    x1=np.linspace(-2,2,1100)
    plt.plot(standardized_x[t==1,0],standardized_x[t==1,1],'o')
    plt.plot(standardized_x[t==0,0],standardized_x[t==0,1],'^')
    plt.plot(x1,-(parameter[0]+parameter[1]*x1)/parameter[2],linestyle='solid')
    plt.grid(True)
    plt.show()
    
    input_x1=float(input('x1を入力してください=> '))
    input_x2=float(input('x2を入力してください=> '))
    input_x=np.array([[input_x1,
                      input_x2]])
    print(1) if sigmoid(create_matrix(input_x),parameter)>=0.5 else print(0)
```

```python
#多項式回帰で最適化を行う
import numpy as np
import matplotlib.pyplot as plt

def standardize(x):
    '''
    データを標準化する
    '''
    x_mean=x.mean(axis=0)
    std=x.std(axis=0)
    return (x-x_mean)/std

def sigmoid(x,parameter):
    return 1/(1+np.exp(-np.dot(x,parameter)))

def create_matrix(x):
    x0=np.ones([x.shape[0],1])#x.shape[0] 配列ｘの行数
    x3=x[:,0,np.newaxis]**2
    return np.hstack([x0,x,x3])#横方向に行列を連結する [1,x1,x2]

def logistic_regression(matrix_x,t):
    '''
    ロジスティック回帰で最適化を行う
    '''
    parameter=np.random.rand(4)
    
    LNR=1e-3
    loop=3000
    count=0
    
    for i in range(loop):
        parameter=parameter-LNR*(np.dot((sigmoid(matrix_x,parameter)-t),matrix_x))
        count+=1
        print('({}) parameter: {}'.format(count,parameter))
    
    return parameter
    
###実行ブロック###
if __name__=='__main__':
    data=np.loadtxt('classification2.csv',dtype='float',delimiter=',',skiprows=1)
    x=data[:,0:2]
    t=data[:,2]
    
    #標準化
    standardized_x=standardize(x)
    
    #行列ｘの生成
    matrix_x=create_matrix(standardized_x)
    
    #学習
    parameter=logistic_regression(matrix_x,t)
    
    #グラフの表示
    x0=np.linspace(-2,2,1100)
    plt.plot(standardized_x[t==1,0],standardized_x[t==1,1],'o')
    plt.plot(standardized_x[t==0,0],standardized_x[t==0,1],'^')
    plt.plot(x0,-(parameter[0]+parameter[1]*x0+parameter[3]*x0**2)/parameter[2],linestyle='solid')
    plt.grid(True)
    plt.show()
    
    input_x1=float(input('x1を入力してください=> '))
    input_x2=float(input('x2を入力してください=> '))
    input_x=np.array([[input_x1,
                      input_x2]])
    print(1) if sigmoid(create_matrix(input_x),parameter)>=0.5 else print(0)
```

```python
#多項式回帰を確率的勾配降下法で行う
import numpy as np
import matplotlib.pyplot as plt

def standardize(x):
    '''
    データを標準化する
    '''
    x_mean=x.mean(axis=0)
    std=x.std(axis=0)
    return (x-x_mean)/std

def sigmoid(x,parameter):
    return 1/(1+np.exp(-np.dot(x,parameter)))

def create_matrix(x):
    x0=np.ones([x.shape[0],1])#x.shape[0] 配列ｘの行数
    x3=x[:,0,np.newaxis]**2
    return np.hstack([x0,x,x3])#横方向に行列を連結する [1,x1,x2]

def logistic_polynomial_stochastic(matrix_x,t):
    '''
    多項式回帰で最適化を行う
    '''
    parameter=np.random.rand(4)
    
    LNR=1e-3
    loop=3000
    count=0
    print(matrix_x)
    for i in range(loop):
        #確率的勾配降下法でパラメータを最適化する
        p=np.random.permutation(matrix_x.shape[0])
        
        for tmp_x,tmp_t in zip(matrix_x[p,:],t[p]):
            parameter=parameter-LNR*(np.dot((sigmoid(tmp_x,parameter)-tmp_t),tmp_x))
        count+=1
        print('({}) parameter: {}'.format(count,parameter))
    
    return parameter
    
###実行ブロック###
if __name__=='__main__':
    data=np.loadtxt('classification2.csv',dtype='float',delimiter=',',skiprows=1)
    x=data[:,0:2]
    t=data[:,2]
    
    #標準化
    standardized_x=standardize(x)
    
    #行列ｘの生成
    matrix_x=create_matrix(standardized_x)
    
    #学習
    parameter=logistic_polynomial_stochastic(matrix_x,t)
    
    #グラフの表示
    x0=np.linspace(-2,2,1100)
    plt.plot(standardized_x[t==1,0],standardized_x[t==1,1],'o',label='1')
    plt.plot(standardized_x[t==0,0],standardized_x[t==0,1],'^',label='0')
    plt.plot(x0,-(parameter[0]+parameter[1]*x0+parameter[3]*x0**2)/parameter[2],linestyle='solid')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    input_x1=float(input('x1を入力してください=> '))
    input_x2=float(input('x2を入力してください=> '))
    input_x=np.array([[input_x1,
                      input_x2]])
    print(1) if sigmoid(create_matrix(input_x),parameter)>=0.5 else print(0)
```
